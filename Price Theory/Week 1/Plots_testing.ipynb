```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of x and y values
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Utility function
def utility(x, y):
    return x**0.5 + y**0.5

# Create the plots
plt.figure(figsize=(12, 10))

# Scenario 1: Both products are good
Z1 = utility(X, Y)
plt.subplot(2, 2, 1)
plt.contourf(X, Y, Z1, levels=20, cmap='coolwarm')
plt.title('Both Products are Good')
plt.xlabel('Product X')
plt.ylabel('Product Y')

# Scenario 2: Both products are bad
Z2 = -utility(X, Y)
plt.subplot(2, 2, 2)
plt.contourf(X, Y, Z2, levels=20, cmap='coolwarm')
plt.title('Both Products are Bad')
plt.xlabel('Product X')
plt.ylabel('Product Y')

# Scenario 3: Product Y is good, Product X is bad (Convex)
Z3 = utility(Y, 1) - utility(X, 1)
plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z3, levels=20, cmap='coolwarm')
plt.title('Product Y is Good, Product X is Bad')
plt.xlabel('Product X')
plt.ylabel('Product Y')

# Scenario 4: Product X is good, Product Y is bad (Convex)
Z4 = utility(1, X) - utility(1, Y)
plt.subplot(2, 2, 4)
plt.contourf(X, Y, Z4, levels=20, cmap='coolwarm')
plt.title('Product X is Good, Product Y is Bad')
plt.xlabel('Product X')
plt.ylabel('Product Y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for x and y
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)

# Create a meshgrid from x and y values
X, Y = np.meshgrid(x, y)

# Compute the function min(x, y)
Z = np.minimum(X, Y)

# Create a filled contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour, label='min(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Indifference Curves: min(x, y)')
plt.grid(True)
plt.show()
```
