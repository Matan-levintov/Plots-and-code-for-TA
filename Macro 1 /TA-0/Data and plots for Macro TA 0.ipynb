```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Download historical data for S&P 500 from Yahoo Finance
sp500 = yf.Ticker(ticker="^GSPC")
# Set the time frame for historical data (adjust as needed)
start_date = "1931-1-1"
end_date = "2020-12-31"

# Fetch historical data
data = sp500.history(period="1d", start=start_date, end=end_date)

# Resample data to have 10-year time skips
resampled_data = data.resample(rule="10Y").last()

# Create a line plot using seaborn
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.lineplot(x=resampled_data.index, y=resampled_data["Close"])
plt.title("S&P 500 Historical Data with 10-Year Time Skips")
plt.xlabel(xlabel="Year")
plt.ylabel(ylabel="S&P 500 Close Price")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
resampled_data['ln_close'] = resampled_data["Close"].apply(lambda x :np.log(x))
# Create a line plot using seaborn
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.lineplot(x=resampled_data.index, y=resampled_data["ln_close"])
plt.title("LOG S&P 500 Historical Data with 10-Year Time Skips")
plt.xlabel("Year")
plt.ylabel("")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np 

data = pd.read_excel("US-ISRAEL-GDP data.xlsx")
#transpose the data so it's easier to work with
transposed_data = data.transpose().reset_index()
transposed_data.columns = ['Year','Israel', 'US']
#Delete the first 4 rows (0,1,2,3)
transposed_data = transposed_data[4:].reset_index()
# transposed_data['ln_Israel'] = np.log(transposed_data['Israel'])
transposed_data['ln_Israel'] = transposed_data['Israel'].apply(lambda x : np.log(x))
transposed_data['ln_US'] = transposed_data['US'].apply(lambda x : np.log(x))




plt.figure(figsize=(10, 6))
sns.lineplot(data=transposed_data, x ="Year",y ='ln_Israel',label = 'Israel')
sns.lineplot(data=transposed_data, x ="Year",y ='ln_US',label = 'US')
plt.title("LOG GDP of Israel and US")
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=transposed_data, x ="Year",y ='Israel',label = 'Israel')
sns.lineplot(data=transposed_data, x ="Year",y ='US',label = 'US')
plt.title("GDP of Israel and US")
plt.ylabel('')
plt.show()
```
